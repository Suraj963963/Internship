```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.dineout.co.in/delhi-restaurants/buffet-special'

page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')
soup

page

soup = BeautifulSoup(page.content) 

first_tital = soup.find('a',class_="restnt-name ellipsis")
first_tital

first_tital.text

loc = soup.find('div',class_="restnt-loc ellipsis")
loc.text

sta = soup.find('span',class_="double-line-ellipsis")
sta.text

sta = soup.find('span',class_="double-line-ellipsis")
sta.text.split('|')[0]

titles = []
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    titles.append(i.text.split('|')[0])
      
titles
    

cuisines = []
for i in soup.find_all('span',class_="double-line-ellipsis"):
        cuisines.append(i.text.split('|')[-1])
    
cuisines


location = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
    
location
    

ratings = []
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    ratings.append(i.string)
    
ratings

price = []
for i in soup.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text.split('|')[0])

price

image = []
for i in soup.find_all("img",class_="no-img"):
    image.append(i.get("data-src"))

image

print(len(titles),len(cuisines),len(location),len(ratings),len(price),len(image))

import pandas as pd
df = pd.DataFrame({'Titles':titles,'Cuisines':cuisines,'Location':location,'Ratings':ratings,'Price':price,'Image_url':image})
df

```
